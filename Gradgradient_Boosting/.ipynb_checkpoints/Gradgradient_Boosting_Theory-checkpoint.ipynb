```python
import numpy as np
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
%pylab inline
```

```python
def L(y, z):
    return z - y
```

```python
def gbm_predict(X):
    '''
    прогноз построенной на данный момент композиции деревьев на выборке `X`
    '''
    return [sum([coeff * algo.predict([x])[0] for algo, coeff in zip(base_algorithms_list, coefficients_list)]) for x in X]
```

# Градиентный бустинг

```python
boston = load_boston()  #выгружаем данные
X = boston.data #признаки
Y = boston.target #целевая перемнная

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=51) #разбиение на отложенную и обучающую выборки

print(X.shape, Y.shape) 
```

```python
base_algorithms_list = list()
coefficients_list    = list()
y_cur = np.array(y_train)
for i in range(50):
    regressor = DecisionTreeRegressor(max_depth=5, random_state=42).fit(X_train, y_cur) #решающие деревья
    base_algorithms_list.append(regressor) #добавление алгоритма к базовым алгоритмам
    coefficients_list.append(0.9 / (1.0 + i)) #уменьшаем вес перед каждым алгоритмом с каждой следующей итерацией 
    
    y_cur = L(gbm_predict(X_train), y_train) #отклонение ответа композиции `a(x)` от правильного ответа `y` на данном `x`
    
error = np.sqrt(mean_squared_error(y_test, gbm_predict(X_test))) #значение функции потерь по компазициям на обчуюащей выборке
    

print(error)
```
